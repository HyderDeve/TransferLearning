# Transfer Learning Static Test Appoarch
### This is the hard code version for testing the model built using Transfer Learning

```python
from PIL import Image
import numpy as np
from tensorflow.keras.models import load_model
```

```python
def image_preprocessing(file_path: str):

    img = Image.open(file_path).convert('RGB').resize((150,150))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array
```

```python
def img_show(file_path):
    img = Image.open(file_path)
    img.show()
```

```python
model = load_model('horses_humans_classifier.keras')

#Prediction of the image
preprocessed_img = image_preprocessing('sufi.png')
prediction = model.predict(preprocessed_img)

print('Given Image is: human' if prediction > 0.5 else 'Given Image is: horse')
```

```python
#Now we confirm whether the prediction is correct or not 
img_show('sufi.png')
```
